```python
import timeit
import random
import matplotlib.pyplot as plt

# Lists to store the values
sizes = []
list_times = []
dict_times = []

for i in range(10000, 1000001, 20000):
    t = timeit.Timer("random.randrange(%d) in x" % i, "from __main__ import random, x")
    
    # Timing for list
    x = list(range(i))
    lst_time = t.timeit(number=1000)
    
    # Timing for dictionary
    x = {j: None for j in range(i)}
    d_time = t.timeit(number=1000)
    
    # Store the results
    sizes.append(i)
    list_times.append(lst_time)
    dict_times.append(d_time)

    # Print the results as before
    print("%d, %10.3f, %10.3f" % (i, lst_time, d_time))

# Plotting the results
plt.plot(sizes, list_times, label="List")
plt.plot(sizes, dict_times, label="Dictionary")
plt.xlabel('Size of the data structure')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of List vs Dictionary Lookup')
plt.legend()
plt.grid(True)
plt.show()
```
